```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize input data
x1, x2 = 0.5, 0.8  # Inputs
y_true = 1  # Desired output (ground truth)

# Initialize weights, bias, and learning rate
w1, w2, b = np.random.randn(), np.random.randn(), np.random.randn()
learning_rate = 0.1  # You can experiment with different values

# Define the sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the derivative of the sigmoid function
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Training loop
errors = []  # List to store error values for each iteration
iterations = 50  # Total number of iterations

print(f"Initial weights: w1={w1:.4f}, w2={w2:.4f}, bias={b:.4f}")

for i in range(iterations):
    # Forward pass
    z = w1 * x1 + w2 * x2 + b  # Weighted sum
    y_pred = sigmoid(z)  # Output after applying activation function

    # Compute the error
    error = 0.5 * (y_true - y_pred) ** 2  # Mean Squared Error (MSE)
    errors.append(error)

    # Backward pass: Compute gradients (TODO placeholders)
    # Derivative of error with respect to output (y_pred)
    dE_dy = y_pred - y_true
    # Derivative of output (y_pred) with respect to z
    dy_dz = sigmoid_derivative(z)
    # Derivative of z with respect to weights and bias
    dz_dw1 = x1  # Input x1 affects w1
    dz_dw2 = x2  # Input x2 affects w2
    dz_db = 1    # Bias term

    # Gradients (TODO: Replace these placeholders with actual calculations)
    dE_dw1 = 0.0  # Replace 0.0 with: dE_dy * dy_dz * dz_dw1
    dE_dw2 = 0.0  # Replace 0.0 with: dE_dy * dy_dz * dz_dw2
    dE_db = 0.0   # Replace 0.0 with: dE_dy * dy_dz * dz_db

    # Update weights and bias
    w1 -= learning_rate * dE_dw1
    w2 -= learning_rate * dE_dw2
    b -= learning_rate * dE_db

    # Print updates every 10 iterations
    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}: w1={w1:.4f}, w2={w2:.4f}, bias={b:.4f}, error={error:.4f}")

# Plot the error reduction over iterations
plt.figure(figsize=(8, 5))
plt.plot(errors)
plt.title("Error Reduction Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.grid()
plt.show()

print("Final weights and bias:")
print(f"w1={w1:.4f}, w2={w2:.4f}, bias={b:.4f}")
```
