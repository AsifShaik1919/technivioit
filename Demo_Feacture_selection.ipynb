```python
import pandas as pd

dataset = pd.read_csv('/content/data.csv')

dataset.head()
```

```python
len(dataset.columns)
```

```python
dataset = dataset.drop(columns=['Unnamed: 32', 'id'])
```

```python
dataset.columns
```

```python
X = dataset[['radius_mean', 'texture_mean', 'area_mean', 'smoothness_mean',
         'compactness_mean', 'concavity_mean', 'concave points_mean',
          'fractal_dimension_mean', 'symmetry_mean']]

y = dataset['diagnosis']
```

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_classif
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
k_best = SelectKBest(score_func=f_classif, k=5)
X_train_k_best = k_best.fit_transform(X_train, y_train)


feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': k_best.scores_, 'p-values': k_best.pvalues_})
print(feature_scores)

X_test_k_best = k_best.transform(X_test)
```

```python
X_k_best = pd.DataFrame(X_train_k_best, columns=X.columns[k_best.get_support()])
X_k_best.head()
```

```python
percentile_selector = SelectPercentile(score_func=f_classif, percentile=60)
X_train_percentile = percentile_selector.fit_transform(X_train, y_train)

feature_scores_percentile = pd.DataFrame({'Feature': X.columns, 'Score': percentile_selector.scores_, 'p-values': percentile_selector.pvalues_})
print(feature_scores_percentile)

X_test_percentile = percentile_selector.transform(X_test)
```

```python
X_train_percentile = pd.DataFrame(X_train_percentile, columns=X.columns[percentile_selector.get_support()])
X_train_percentile.head()
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
## K best features

model = MLPClassifier(max_iter=1000, random_state=42)
model.fit(X_k_best, y_train)

accuracy = model.score(X_test_k_best, y_test)
print(f"Accuracy: {accuracy}")
```

```python
## Select Percentile

model = MLPClassifier(max_iter=1000, random_state=42)
model.fit(X_train_percentile, y_train)

accuracy = model.score(X_test_percentile, y_test)
print(f"Accuracy: {accuracy}")
```
